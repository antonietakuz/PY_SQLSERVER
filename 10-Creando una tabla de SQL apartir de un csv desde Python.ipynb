#### 3. Trabajamos con SQL

```python
import pyodbc
```

```python
#Creamos una variable donde guardamos los datos de conexion del servidor SQL
server ="DESKTOP-GU4H9H0\SQLEXPRESS"
```

```python
conexion =pyodbc.connect(driver='{SQL Server}', host =server, database = "DB_PYTHON")
print("conexion exitosa")
```

```python
#Creamos un cursor para almacenar la informacion en memoria
cursor= conexion.cursor()
```

```python
#Indicamos lo que queremos que se lleve a cabo en memoria

cursor.execute('''
CREATE TABLE MaratonNY_Python (
    corredor_id int primary key,
    Lugar int,
    Genero varchar(25),
    Edad int,
    Pais varchar(255),
    Tiempo float)
               ''')
conexion.commit()
```

```python
# PREPARED STATEMENT (NO DATA)
query = "insert into MaratonNY_Python (corredor_id, Lugar, Genero, Edad,Pais,Tiempo) values (?,?,?,?,?,?)"
```

```python

# EXECUTE STATEMENT AND BIND PARAMS
cursor.executemany(query, lista_valores)
cursor.commit()
```
